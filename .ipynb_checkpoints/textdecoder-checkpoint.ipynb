```python
#FOR DECODE. Create decimal from padded binary
def largenumdec(): #using double Size Method
    global beginbin
    oneindex = beginbin.index('1')
    numlength = oneindex * 2 + 1
    print(''.join(beginbin[:numlength]))
    #assert 0 < oneindex
    assert len(beginbin) >= numlength-1 #+1-1
    toret = int("".join(beginbin[oneindex:numlength]),2)
    beginbin = beginbin[numlength:]
    return toret
```

```python
#v1.1, 5 bits per character, numbers have an added 5 bits.
global workingstr
global beginbin
while len(beginbin) > 0:
    assert len(beginbin) >= 5
    print(''.join(beginbin))
    #currentbit = beginbin.pop(0)
    #print(currentbit)
    incharnum = int("".join(beginbin[:5]),2)
    if(verb >= 2):
        print(incharnum)
    beginbin = beginbin[5:]
    if(incharnum < 26): #h
        if(verb >= 1):
            print('is letter')
        inchar = chr(incharnum + 97)
        print(inchar)
        workingstr.append(inchar)
    elif(incharnum == 26):
        if(verb >= 1):
            print('is space')
        workingstr.append(' ')
    elif(incharnum == 27):
        if(verb >= 1):
            print('is number')
        innum = largenumdec()
        print(innum)
        workingstr += list(str(innum))
    else:
        if(verb >= 1):
            print('unrecognized binary at bit ' + str(len(inbinary)-(len(beginbin) + 5)) + '.')
        assert False
print("Done!")
```

```python
def retrieve(retlen):
    if verb >= 2:
        print('retrieving ' + str(retlen))
    assert len(beginbin) >= retlen
    toret = []
    for i in range(retlen):
        toret.append(beginbin.pop(0))
    return join(toret)
```

```python
def alphanumeric(inbin):
    if(verb >= 1):
        print('alphanumeric')
    #1 bit
    if(inbin[0] == '0'):
        #3 bits 0-25 is a-z
        incharnum = int(inbin[1:4],2)
        if(incharnum < 26): #h
            if(verb >= 1):
                print('is letter')
            inchar = chr(incharnum + 97)
            print(inchar)
            workingstr.append(inchar)

    elif(inchar[0] == '1'):
        print('1')

    return 
```
